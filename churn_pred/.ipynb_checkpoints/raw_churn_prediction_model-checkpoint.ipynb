#### 4.0 Data split & Metrics setting for cross-validation

```python
metric_list = ['accuracy', 'precision', 'recall', 'f1_score']
```

```python
def matrix_to_metrics(mat):
    TP = mat[1, 1]
    TN = mat[0, 0]
    FP = mat[0, 1]
    FN = mat[1, 0]
    
    acc = (TP + TN) / (TP + TN + FP + FN)
    prec = TP / (TP + FP) 
    rec = TP / (TP + FN)
    f1_score = 2 * (prec * rec) / (prec + rec)
    
    return [acc, prec, rec, f1_score]
```

```python
from sklearn.compose import ColumnTransformer
```

```python
def cross_validate(model, data, scaling, pass_cols, cols_to_drop):

    result_dict = {} 
    
    metrics = ['accuracy', 'precision', 'recall', 'f1_score']
    for m in metrics:
        result_dict[m] = []

    X_train, y_train = data
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    
    # removing variables with multicollinearity
    if cols_to_drop is not None:
        X_train = X_train.drop(columns=cols_to_drop, axis=1)
        print(f'# of colums after drop: {len(X_train.columns)}')

    bin_cols = ['Complains', 'Tariff Plan', 'Status']
    num_cols = [c for c in X_train.columns if c not in bin_cols and c not in pass_cols]
    print(num_cols)
    check_scaling = True
    
    start_time = time.time()
    for fold in kf.split(X_train): 
        trn_idx, val_idx = fold
        # training data split for analysis & assessment
        X_trn, X_val = X_train.iloc[trn_idx], X_train.iloc[val_idx] # analysis set 
        y_trn, y_val = y_train.iloc[trn_idx], y_train.iloc[val_idx] # assessment set
        # scaling
        if scaling:
            preprocessor = ColumnTransformer(
                transformers=[
                    ('num', StandardScaler(), num_cols), 
                    ('binary', 'passthrough', bin_cols),
                    ('pca-ed', 'passthrough', [c for c in X_train.columns if 'PC' in c])
                ]
            )
            if check_scaling is True:
                print(preprocessor)
                check_scaling = False
                
            X_trn = preprocessor.fit_transform(X_trn)  # Transformation of X_trn using learned parameters
            X_val = preprocessor.transform(X_val)  # Transformation of X_val with the same parameters
        # model fitting
        model.fit(X_trn, y_trn)
        # model inference
        y_pred = model.predict(X_val)
        # model evaluation
        conf_mat = confusion_matrix(y_val, y_pred)
        fold_result = matrix_to_metrics(conf_mat)
        for i, val in enumerate(fold_result):
            result_dict[metrics[i]].append(round(fold_result[i], 4))
            
    end_time = time.time() 
    elapsed_time = end_time - start_time
    readable_time = f"{elapsed_time:.3f} seconds"
    print(readable_time)

    for key in result_dict:
        result_dict[key].append({'average': round(sum(result_dict[key]) / len(result_dict[key]), 3)})
    
    return result_dict
```

```python
def get_summary(res_dict):
    cols = []
    avgs = []
    for key in res_dict:
        cols.append(key)
        avgs.append(res_dict[key][-1]['average'])
    return pd.DataFrame({'Metric': cols, 'Average': avgs})       
```

```python
def compare_models(df_dict): 
    n = len(df_dict)
    for idx, items in enumerate(df_dict.items()):
        key, df = items
        if idx == 0:
            merged_df = df_dict[key].rename(columns={'Average': key})
        else:
            merged_df = pd.merge(merged_df, df.rename(columns={'Average': key}), on='Metric')

    merged_df = merged_df.set_index('Metric')
    merged_df['best_model'] = merged_df.idxmax(axis=1)
    return merged_df
```

```python
def init_model():
    model_dict = {}
    model_dict['saga'] = LogisticRegression(solver='saga', penalty='none', max_iter=200)
    model_dict['l1_saga'] = LogisticRegression(solver='saga', penalty='l1', max_iter=200)
    model_dict['l2_saga'] = LogisticRegression(solver='saga', max_iter=200)
    model_dict['lbfgs'] = LogisticRegression(penalty='none', max_iter=200)
    model_dict['l2_lbfgs'] = LogisticRegression(max_iter=200) # default    
    return model_dict    
```

```python
# data, model, scaling=False, pass_cols=[], cols_to_drop=None
def run(param_dict):
    result_dict = {}
    model_dict = init_model()
    for name in model_dict:
        print(name, '='* (20-len(name)))
        model = model_dict[name]
        single_result = cross_validate(model, data=param_dict['data'], scaling=param_dict['scaling'], 
                                       pass_cols=param_dict['pass_cols'], cols_to_drop=param_dict['cols_to_drop'])
        result_dict[name] = get_summary(single_result)   
    return result_dict, model_dict
```

```python
# check variable coefficient
def show_coef(model_dict, columns):    
    coef_df = pd.DataFrame(index=columns)
    
    for key in model_dict:
        model = model_dict[key]
        coefficients = np.round(model.coef_[0], decimals=4)  # 각 피처의 계수
        coef_df[key] = coefficients  # 모델 이름을 열로 추가

    avgs = []
    for c in columns:
        avgs.append(round(coef_df.loc[c].mean(), 4))

    coef_df['average'] = avgs
        
    return coef_df
```

#### 4.2.1 Applying PCA only to continuous features with high correlations

```python
from sklearn.decomposition import PCA

def apply_PCA(data, num_comp):
    scaler = StandardScaler()
    scaled_data = scaler.fit_transform(data)
    pca = PCA(n_components=num_comp)
    principal_components = pca.fit_transform(scaled_data)
    cols = [f'PC{i}' for i in range(1, num_comp+1)]
    pca_df = pd.DataFrame(data=principal_components, columns=cols)
    return pca_df
```

### 4.4 Neural Net

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset, Dataset

from sklearn.base import BaseEstimator, ClassifierMixin
```

```python
class BinaryClassificationNet(nn.Module):
    def __init__(self, input_dim):
        super(BinaryClassificationNet, self).__init__()
        self.fc1 = nn.Linear(input_dim, 8)
        self.fc2 = nn.Linear(8, 1)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        return x
```

```python
class PytorchModelWrapper(BaseEstimator, ClassifierMixin):
    def __init__(self, model, input_dim):
        self.model = model
        self.epochs = 50
        self.lr = 0.001
        self.criterion = nn.BCEWithLogitsLoss()
        self.optimizer = torch.optim.Adam(model.parameters(), lr=self.lr)
        
    def fit(self, X, y):
        X_tensors = torch.tensor(X, dtype=torch.float32)
        y_tensors = torch.tensor(y.to_numpy(), dtype= torch.float32).view(-1, 1)

        train_dataset = TensorDataset(X_tensors, y_tensors)
        train_loader = DataLoader(dataset=train_dataset, batch_size=32, shuffle=True)

        self.model.train()
        for epoch in range(self.epochs):
            epoch_loss = 0
            for batch in train_loader:
                x_batch, y_batch = batch
                self.optimizer.zero_grad()
                outputs = self.model(x_batch)
                loss = self.criterion(outputs, y_batch)
                loss.backward()
                self.optimizer.step() 
                epoch_loss += loss.item()
            # print log
            print(f'Epoch [{epoch+1}/{self.epochs}], Loss: {epoch_loss/len(train_loader):.4f}')
            
    def predict(self, X):
        X_tensors = torch.tensor(X, dtype=torch.float32)
        val_dataset = TensorDataset(X_tensors)
        val_dataloader = DataLoader(val_dataset, batch_size=32, shuffle=False)
    
        predictions = []
        
        self.model.eval()
        with torch.no_grad():
            for batch in val_dataloader:
                outputs = self.model(batch[0])
                pred = (outputs > 0.5).float().numpy().ravel() 
                predictions.extend(pred)

        return np.array(predictions)
```

```python
def run_nn():
    result_dict = {}
    input_dim = len(X_train.columns)
    model = BinaryClassificationNet(input_dim)
    nn_model = PytorchModelWrapper(model, input_dim)    
   
    print('Neural Net', '='*(20-len(key)))
    key_df = cross_validate([X_train, y_train], nn_model, scaling=True) #cols_to_drop=None
    result_dict['Neural Net'] = get_summary(key_df)    
        
    return result_dict, nn_model
```

```python
def hist_comp_trans(col_array, trans, trans_df):
    n = len(col_array)
    fig, axes = plt.subplots(n, 2, figsize=(8, 4 * n))
    for i, col in enumerate(col_array): 
        sns.histplot(data=X_train, x=col, ax=axes[i, 0])
        axes[i, 0].set_title(f'Original {col}')
        
        # Log-transformed data histogram (add 1 to avoid log(0))
        sns.histplot(data=trans_df, x=col, ax=axes[i, 1], color='lightgreen')
        axes[i, 1].set_title(f'{trans}-Transformed {col}')
        
        for ax in axes[i]:
            ax.set_xlabel('Value')
            ax.set_ylabel('Frequency')
        
    # Show the histograms
    plt.tight_layout()
    plt.show()
```

```python
hist_comp_trans(continuous_features, 'Log', log_transformed)
```

```python
# Skewness comparison before and after transformation
from scipy.stats import skew
def get_skewness(df):
    cols = df.columns
    keys = ['col_name', 'before', 'after', 'diff']
    result = {k:[] for k in keys}
    for c in cols:
        result['col_name'].append(c)
        pre = skew(X_train[c])
        post = skew(df[c])
        diff = abs(pre) - abs(post)
        result['before'].append(pre)
        result['after'].append(post)
        result['diff'].append(diff)
    res_df = pd.DataFrame(result)
    return res_df
```

#### ?.2 Yeo-Johnson Transformation 
- Yeo-Johnson for predictors with zeros
- YJ transformation outperforms in reducing skewness across all columns.
- Use YJ-transformed columns when the model requires normally distributed variables.

```python
from sklearn.preprocessing import PowerTransformer
```

```python
def yeo_johnson_transform(df):
    transformer = PowerTransformer(method='yeo-johnson', standardize=False)
    df_transformed = df.copy()
    df_transformed[df.columns] = transformer.fit_transform(df)
    return df_transformed
```

### 5. Adjustment for class imbalance
- Model fixed to the best performing LightGBM from CV

```python
from imblearn.over_sampling import SMOTE

def cross_validate(data=[X_train, y_train], weight=False, auto=False, oversampling=False, curve=False):

    result_dict = {} 
    
    metrics = ['accuracy', 'precision', 'recall', 'f1_score']
    for m in metrics:
        result_dict[m] = []

    X_train, y_train = data
    kf = KFold(n_splits=5, shuffle=True, random_state=42)

    tprs = []
    aucs = []
    mean_fpr = np.linspace(0, 1, 100)
    plt.figure(figsize=(8, 6))
    
    start_time = time.time()
    for fold_idx, fold in enumerate(kf.split(X_train)): 
        trn_idx, val_idx = fold
        # training data split for analysis & assessment
        X_trn, X_val = X_train.iloc[trn_idx], X_train.iloc[val_idx] # analysis set 
        y_trn, y_val = y_train.iloc[trn_idx], y_train.iloc[val_idx] # assessment set
        
        # weight
        if weight is True:
            scale_pos_weight = len(y_trn[y_trn == 0]) / len(y_trn[y_trn == 1])
            model = LGBMClassifier(scale_pos_weight=scale_pos_weight)
            
        # oversampling 
        if oversampling is True:
            smote = SMOTE()
            X_trn, y_trn = smote.fit_resample(X_trn, y_trn)
            model = LGBMClassifier()

        # auto
        if auto is True:
            model = LGBMClassifier(is_unbalance=True)
            
        # model fitting
        model.fit(X_trn, y_trn)
        # model inference
        y_pred = model.predict(X_val)
        # model evaluation
        conf_mat = confusion_matrix(y_val, y_pred)
        fold_result = matrix_to_metrics(conf_mat)
        for i, val in enumerate(fold_result):
            result_dict[metrics[i]].append(round(fold_result[i], 4))

        # roc-auc curve
        if curve is True:
            y_pred_proba = model.predict_proba(X_val)[:, 1]
            
            # ROC 곡선과 AUC 계산
            fpr, tpr, thresholds = roc_curve(y_val, y_pred_proba)
            roc_auc = auc(fpr, tpr)
            aucs.append(roc_auc)
            
            # ROC 커브 그리기
            plt.plot(fpr, tpr, lw=2, alpha=0.3, label=f'Fold {fold_idx+1} (AUC = {roc_auc:.2f})')
        
            # 보간법으로 TPR 저장 (평균 ROC 계산에 사용)
            tprs.append(np.interp(mean_fpr, fpr, tpr))
            tprs[-1][0] = 0.0

            # 최적 임계값 선택 (예시로 TPR과 FPR의 비율을 고려하여 선택)
            optimal_idx = np.argmax(tpr - fpr)
            optimal_threshold = thresholds[optimal_idx]
            print(f"\nOptimal threshold based on ROC curve: {optimal_threshold:.2f}")
            
            # 최적 임계값으로 예측
            y_pred_optimal = (y_pred_proba >= optimal_threshold).astype(int)
            print("\nClassification report with optimal threshold:\n")
            print(classification_report(y_val, y_pred_optimal))
                        
    end_time = time.time() 
    elapsed_time = end_time - start_time
    readable_time = f"{elapsed_time:.3f} seconds"
    print(readable_time)

    # 평균 ROC 곡선 및 AUC 계산
    mean_tpr = np.mean(tprs, axis=0)
    mean_tpr[-1] = 1.0
    mean_auc = auc(mean_fpr, mean_tpr)
    std_auc = np.std(aucs)
    
    plt.plot(mean_fpr, mean_tpr, color='b', label=r'Mean ROC (AUC = %0.2f $\pm$ %0.2f)' % (mean_auc, std_auc), lw=2)
    
    # 대각선 기준선 그리기
    plt.plot([0, 1], [0, 1], linestyle='--', lw=2, color='r', alpha=0.8)
    
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC-AUC Curves for Each Fold')
    plt.legend(loc="lower right")
    plt.show()

    
    for key in result_dict:
        result_dict[key].append({'average': round(sum(result_dict[key]) / len(result_dict[key]), 3)})
    
    return result_dict
```
