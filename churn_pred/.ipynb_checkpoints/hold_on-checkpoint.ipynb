```python
def run_total(data, model_class, model_params, cust_scaler=None):
    result_dict = {} 

    X_train, y_train, X_valid, y_valid = data
    
    # removing correlated columns
    X_train.drop(columns=params_dict['drop'], axis=1, inplace=True)

    start_time = time.time()
  
    # scaling
    if cust_scaler is not None:                
        X_train = cust_scaler.fit_transform(X_train)  

    # model fitting
    model = model_class(**model_params)        
    model.fit(X_train, y_train)
    
    # model evaluation
    y_pred = model.predict(X_valid)
    conf_mat = confusion_matrix(y_valid, y_pred)
    [acc, prec, rec, f1] = matrix_to_metrics(conf_mat)
    
    result_dict['accuracy'] = acc
    result_dict['precision'] = prec
    result_dict['recall'] = rec
    result_dict['f1_score'] = f1
    
            
    end_time = time.time() 
    elapsed_time = end_time - start_time
    readable_time = f"{elapsed_time:.3f} seconds"
    print(readable_time)
    
    return result_dict, model
```

```python
def run_final(data, model):
    result_dict = {} 

    y_pred = model.predict(X_valid)
    conf_mat = confusion_matrix(y_valid, y_pred)
    [acc, prec, rec, f1] = matrix_to_metrics(conf_mat)
    
    result_dict['accuracy'] = acc
    result_dict['precision'] = prec
    result_dict['recall'] = rec
    result_dict['f1_score'] = f1

    return result_dict   
    
```
